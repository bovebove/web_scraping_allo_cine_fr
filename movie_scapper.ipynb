### Scrapper la premiere page

```python
#liste des urls
urls=[
    'https://www.allocine.fr/film/meilleurs/',
]
```

```python
urls[0]
```

```python
# extraire les titres des films de la page 1
response=requests.get(urls[0])
soup=BeautifulSoup(response.text,'html.parser')
titres_balise=soup.find_all("a",{"class":"meta-title-link"})
titres=[x.get_text() for x in titres_balise]
titres
```

```python
# extraire les resumés des films de la page 1
response=requests.get(urls[0])
soup=BeautifulSoup(response.text,'html.parser')
resume_balise=soup.find_all("div",{"class":"content-txt"})
resume=[x.get_text() for x in resume_balise]
resume
```

```python
# extraire les notes des films de la page 1
response=requests.get(urls[0])
soup=BeautifulSoup(response.text,'html.parser')
note_balise=soup.find_all("span",{"class":"stareval-note"})
note=[x.get_text() for x in note_balise]
note
```

```python
# reorganiser les notes en trois colonnes
np_array=np.array(note)
reshaped_array=np.reshape(np_array,(len(titres),3))
note_dataframe=pd.DataFrame(reshaped_array,columns=['note_presse','note_spectateur','note_amis'])
note_dataframe
```

```python
#mettre toutes les informations en dataframe
df=pd.DataFrame()
df['titre']=titres
df['resume']=resume
df
```

```python
df=pd.concat([df,note_dataframe],axis=1)
df
```

### Scrapper les autres pages

```python
#creer une fonction qui va generer des liens
def creer_lien(urls):
  for i in range(2,31):
    urls.append(f'https://www.allocine.fr/film/meilleurs/?page={i}')
  return urls
```

```python
creer_lien(urls)
```

```python
# extraire les titres des films de toutes les pages
titre_total=[]
for url in urls:
  response=requests.get(url)
  soup=BeautifulSoup(response.text,'html.parser')
  titres_balise=soup.find_all("a",{"class":"meta-title-link"})
  titres=[x.get_text() for x in titres_balise]
  titre_total+=titres
titre_total
```

```python
# extraire les resumés des films de toutes les pages
resume_total=[]
for url in urls:
  response=requests.get(url)
  soup=BeautifulSoup(response.text,'html.parser')
  resume_balise=soup.find_all("div",{"class":"content-txt"})
  resume=[x.get_text() for x in resume_balise]
  resume_total+=resume
resume_total
```

```python
# extraire les notes des films de la page 1
note_total=[]
for url in urls:
  response=requests.get(url)
  soup=BeautifulSoup(response.text,'html.parser')
  note_balise=soup.find_all("span",{"class":"stareval-note"})
  note=[x.get_text() for x in note_balise]
  note_total+=note
note_total
```

```python
# combiner les titres et les resumées
df=pd.DataFrame()
df['titre']=titre_total
df['resume']=resume_total
df
```

```python
df.to_csv('movies_data.csv')
```
